## Understanding the prediction problem

#### Identify tables:
The tables in the dataset are claims, policy holders and third parties

#### Definitions:
##### • FNOL: 
It refers to the initial report made to an insurance company when a policyholder experiences a loss, damage, or incident that may lead to a claim.

##### • Estimated Claim Amount: 
It is the insurer’s initial estimate of how much the claim is likely to cost, based on the information available at the time.

##### • Ultimate Claim Amount: 
It represents the actual amount the insurer ends up paying, including all components of the claim

##### • Third-party severity: 
A third party is someone other than the insured. Injury, Property damage or financial loss you cause to someone else (the "third party") in an accident. The third-party severity is categorised into minor, serious and fatal


#### Histogram

```python
#Import Libraries
import pandas as pd # Data processing
import matplotlib.pyplot as plt # Data Visualization
```

```python
#Load the claim dataset
claims = pd.read_csv('Data/claims.csv')
```

```python
# Histogram
values = claims["Ultimate_Claim_Amount"]

plt.hist(values, bins=30)
plt.xlabel("Ultimate Claim Amount")
plt.ylabel("Frequency")
plt.title("Histogram of Ultimate Claim Amount")
plt.savefig('Visualizations/Understanding the prediction problem/Histogram of Ultimate Claim Amount.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Histogram (log scale)


plt.figure(figsize=(8, 5))
plt.hist(claims['Ultimate_Claim_Amount'], bins=50)
plt.xscale('log')

plt.xlabel('Ultimate Claim Amount (log scale)')
plt.ylabel('Frequency')
plt.title('Histogram of Ultimate Claim Amount (Log Scale)')
plt.tight_layout
plt.savefig('Visualizations/Understanding the prediction problem/Histogram of Ultimate Claim Amount(Log Scale).png', dpi=300, bbox_inches='tight')
plt.show()
```

##### Insights from the Log-Scale Histogram
 - Highly Right-Skewed Distribution
 - Most claims cluster roughly between 10² and 10⁴ (hundreds to tens of thousands).
 - A smaller but visible group extends into 10⁴–10⁵ (large losses).
 - Very few claims exceed 10⁵, forming the catastrophic tail.
